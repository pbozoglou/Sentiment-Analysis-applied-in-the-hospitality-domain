## Domain Selection and Dataset Creation

The domain selected for this project is hotel reviews. Specifically, a dataset of reviews for lodgings in Athens city, collected using the Google Places API.

```python
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize, FreqDist, pos_tag
from nltk.stem import WordNetLemmatizer, PorterStemmer
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import emoji
from sklearn import model_selection
from sklearn.preprocessing import LabelEncoder
from sklearn import model_selection
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.svm import SVC

pd.set_option('display.max_colwidth', 15)
```

```python
reviews = pd.read_csv("../datasets/reviews_athens.csv")
reviews.head()
```

```python
reviews_df = reviews.copy()
```

```python
reviews_df.info()
```

**Drop duplicate & missing values (only 1 row will be affected for this dataset):**

```python
reviews_df.drop_duplicates(inplace=True)
reviews_df.dropna(inplace=True)
```

**Convert to *positive* and *negative* sentiment:**

```python
_, ax1 = plt.subplots(nrows=1,ncols=2,figsize=(8,3))

reviews_df.loc[reviews_df.user_rating < 3,'sentiment'] = 'negative'
reviews_df.loc[reviews_df.user_rating > 3,'sentiment'] = 'positive'
reviews_df.drop(index=reviews_df[reviews_df.user_rating==3].index, inplace=True)

t1='Average Rating: {}'.format(round(reviews_df.user_rating.mean(),2))
reviews_df.sentiment.value_counts(True).plot(kind='bar',ax=ax1[0],subplots=True,rot=0)
reviews_df.user_rating.plot(kind='hist',ax=ax1[1],subplots=True)
ax1[0].set_xlabel('')
ax1[0].set_ylabel('%')
ax1[0].set_title('Sentiment Distribution')
ax1[1].set_ylabel('')
ax1[1].set_title('Avg Rating: {}'.format(reviews_df.user_rating.mean().round(1)));
```

- **⚠️ Note:** Positive ratings are dominant (more than 80% of the data)

**Randomly drop positive-sentiment rows to have a 1:1 distribution**

```python
np.random.seed(42)

dx = reviews_df.sentiment.value_counts().positive - reviews_df.sentiment.value_counts().negative
random_positives_index = np.random.choice(reviews_df[reviews_df.sentiment=='positive'].index, dx, replace=False)

dataset = reviews_df.drop(index=random_positives_index)[['review_text','user_rating','sentiment']]
print('Final Dataset:')
dataset
```

**Explore the dataset**

```python
_, ax2 = plt.subplots(nrows=1,ncols=3,figsize=(15,3))

dataset.sentiment.value_counts(True).plot(kind='bar',ax=ax2[0],subplots=True,rot=0)
dataset.review_text.apply(word_tokenize).apply(len).plot(kind='hist',ax=ax2[1],subplots=True)
dataset.user_rating.plot(kind='hist',ax=ax2[2],subplots=True)

ax2[0].set_title('Sentiment Distribution')
ax2[0].set_xlabel('')
ax2[0].set_ylabel('%')
ax2[1].set_title('Avg Tokens: {}'.format(dataset.review_text.apply(word_tokenize).apply(len).mean().round(1)))
ax2[1].set_ylabel('Num of Tokens')
ax2[1].set_xlabel('Num of Reviews')
ax2[2].set_ylabel('')
ax2[2].set_title('Avg Rating: {}'.format(dataset.user_rating.mean().round(1)));
```

## Preprocessing and Text Analysis

**Perform text cleaning: remove noise, handle emojis, stop words, and other irrelevant tokens. [[🔗#1](https://www.geeksforgeeks.org/text-preprocessing-for-nlp-tasks/)] [[🔗#2](https://medium.com/sciforce/text-preprocessing-for-nlp-and-machine-learning-tasks-3e077aa4946e)]**

```python
stopwords_eng = set(stopwords.words('english'))
stopwords_eng.remove('not')
lemmatizer = WordNetLemmatizer()

def clean(txt:str)->str:
    txt = emoji.demojize(txt).lower()
    txt = ' '.join([lemmatizer.lemmatize(w) for w in word_tokenize(txt) if w not in stopwords_eng])
    txt = re.sub(r'[^\w\s]|\d+', '', txt)
    return re.sub(r'\n| +',' ',txt).strip()
```

```python
dataset.review_text.iloc[1262]
```

```python
dataset['review_text_cleaned'] = dataset.review_text.apply(clean)
dataset.review_text_cleaned.iloc[1262]
```

**Average tokens after cleaning**

```python
dataset.review_text.apply(word_tokenize).apply(len).mean(), '->', dataset.review_text_cleaned.apply(word_tokenize).apply(len).mean()
```

**Tokenize and vectorize the data using techniques like TF-IDF or word embeddings (e.g.,
Word2Vec, GloVe).**

```python
dataset['tokens'] = dataset.review_text_cleaned.apply(word_tokenize)
dataset['freqdist'] = dataset.tokens.apply(FreqDist)
dataset.freqdist.sum()
```

```python
dataset.freqdist.sum().plot(50);
```

```python
dataset['pos_tag'] = dataset.tokens.apply(pos_tag)
dataset['pos_tag_freqdist'] = dataset.pos_tag.apply(FreqDist)
dataset.pos_tag_freqdist.iloc[0]
```

```python
all_post_tags = dataset.pos_tag_freqdist.apply(lambda x: [y[1] for y in x.keys()]).apply(FreqDist)
all_post_tags.sum().plot(50);
```

```python
dataset.freqdist.sum()
```

```python
'Number of words:', len(dataset.freqdist.sum())
```

```python
word_features = [i[0] for i in dataset.freqdist.sum().most_common(1500)]
word_features[:5]
```

```python
encoder = LabelEncoder()
Y = encoder.fit_transform(dataset.sentiment)
Y
```

```python
def find_features(message):
    words = word_tokenize(message)
    features = {}
    for word in word_features:
        features[word] = (word in words)

    return features
```

```python
messages = list(zip(dataset.review_text_cleaned,Y))
np.random.shuffle(messages)

featuresets = [(find_features(text), label) for (text, label) in messages]
```

```python
training, testing = model_selection.train_test_split(featuresets, test_size = 0.25, random_state=42)
```

```python
print(len(training))
print(len(testing))
```

## Baseline Model Development

```python
model = SklearnClassifier(SVC(kernel = 'linear'))
model.train(training)

accuracy = nltk.classify.accuracy(model, testing)*100
print("SVC Accuracy: {}".format(accuracy))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Define models to train
names = ["K Nearest Neighbors", "Decision Tree", "Random Forest", "Logistic Regression", "SGD Classifier",
         "Naive Bayes", "SVM Linear"]

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    LogisticRegression(),
    SGDClassifier(max_iter = 100),
    MultinomialNB(),
    SVC(kernel = 'linear')
]

models = zip(names, classifiers)

for name, model in models:
    nltk_model = SklearnClassifier(model)
    nltk_model.train(training)
    accuracy = nltk.classify.accuracy(nltk_model, testing)*100
    print("{} Accuracy: {}".format(name, accuracy))
```

```python
# Ensemble methods - Voting classifier
from sklearn.ensemble import VotingClassifier

names = ["K Nearest Neighbors", "Decision Tree", "Random Forest", "Logistic Regression", "SGD Classifier",
         "Naive Bayes", "SVM Linear"]

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    LogisticRegression(),
    SGDClassifier(max_iter = 100),
    MultinomialNB(),
    SVC(kernel = 'linear')
]

models = list(zip(names, classifiers))

nltk_ensemble = SklearnClassifier(VotingClassifier(estimators = models, voting = 'hard', n_jobs = -1))
nltk_ensemble.train(training)
accuracy = nltk.classify.accuracy(nltk_model, testing)*100
print("Voting Classifier: Accuracy: {}".format(accuracy))
```

```python
# make class label prediction for testing set
txt_features, labels = zip(*testing)

prediction = nltk_ensemble.classify_many(txt_features)
```

```python
# print a confusion matrix and a classification report
print(classification_report(labels, prediction))

pd.DataFrame(
    confusion_matrix(labels, prediction),
    index = [['actual', 'actual'], ['positive', 'negative']],
    columns = [['predicted', 'predicted'], ['positive', 'negative']])
```
