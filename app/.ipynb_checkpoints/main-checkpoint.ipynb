```python
import nltk
from nltk.tree import Tree
from IPython.display import display
import numpy as np
import pandas as pd
pd.set_option('display.max_colwidth', 15)
```

```python
reviews_original = pd.read_csv("../datasets/reviews_athens.csv")
reviews_original.head()
```

```python
reviews_original.info()
```

#### 1. Preprocessing

```python
reviews_original.duplicated().sum()
```

```python
# reviews_original.isnull().sum()
```

```python
reviews_original.dropna(inplace=True)
# reviews_original.isnull().sum()
```

```python
ax1 = reviews_original['review_text'].apply(len).hist(figsize=(4,4))
ax1.set_title('Average Character Length: {}'.format(reviews_original['review_text'].apply(len).mean()),pad=15);
```

```python
ax2 = reviews_original['user_rating'].hist(figsize=(4,4))
ax2.set_title('Average Rating: {}'.format(reviews_original['user_rating'].mean()),pad=15);
```

```python
reviews_original['sentiment'] = 'Neutral'
reviews_original.loc[reviews_original['user_rating'] < 3,'sentiment'] = 'Negative'
reviews_original.loc[reviews_original['user_rating'] > 3,'sentiment'] = 'Positive'
reviews_original.drop(index=reviews_original[reviews_original['sentiment']=='Neutral'].index, inplace=True)

val_counts = reviews_original['sentiment'].value_counts(True)
print(val_counts)
val_counts.plot(kind='bar',figsize=(4,4));
```

```python
#Equalize positive:negative ratio to 1:1 by reducing the positive data at random.
np.random.seed(42)
dx = reviews_original['sentiment'].value_counts()['Positive'] - reviews_original['sentiment'].value_counts()['Negative']
index = np.random.choice(reviews_original[reviews_original['sentiment']=='Positive'].index, dx, replace=False)
reviews_original_new = reviews_original.drop(index=index)
print(reviews_original_new['sentiment'].value_counts())
reviews_original_new['sentiment'].value_counts(True).plot(kind='bar',figsize=(4,4));
```

```python
dataset = reviews_original_new[['review_text','sentiment']]
dataset.head(10)
```

```python
ax3 = dataset['review_text'].apply(len).hist(figsize=(4,4))
ax3.set_title('Average Character Length: {}'.format(dataset['review_text'].apply(len).mean()),pad=15);
```

```python
ax4 = reviews_original_new['user_rating'].hist(figsize=(4,4))
ax4.set_title('Average Rating: {}'.format(reviews_original_new['user_rating'].mean()),pad=15);
```

#### 2. NLTK Experimentation

```python
from nltk.text import Text
from nltk import FreqDist
```

```python
sentence = dataset.loc[dataset.index[1],'review_text']
example = Text(nltk.word_tokenize(sentence))
example
```

```python
token_sequence = sentence.split()
freq_dist = FreqDist(token_sequence)
freq_dist.most_common(10)
```

```python
example.concordance("was")
```

```python
example.common_contexts(["was"])
```

```python
example.dispersion_plot([i for i in freq_dist][:5])
```

```python
from collections import Counter
bag_of_words = Counter(token_sequence)
bag_of_words
```

```python
sentence_bow = {} #create a dictionary

for token in sentence.split():
    wordfreq=sentence.count(token)
    sentence_bow[token]=wordfreq
sorted(sentence_bow.items())
```

```python
df = pd.DataFrame(pd.Series(dict([(token, sentence.count(token)) #έβαλα, αντί για 1 την συχνότητα εμφάνισης
    for token in sentence.split()])), columns=['sent']).T
df
```

```python
sentences = '\n'.join([dataset.loc[dataset.index[i],'review_text'].replace('\n','') for i in range(3)])
sentences
```

```python
corpus = {}
for i, sent in enumerate(sentences.split('\n')):
    corpus['sent{}'.format(i)] = dict((tok.strip('.'), sentence.count(tok)) for tok in sent.split())
    
df = pd.DataFrame.from_records(corpus).fillna(0).astype(int).T
df[df.columns[::]]
```

```python
s1 = sentences.split('\n')[1].lower()
s1
```

```python
from nltk import word_tokenize, pos_tag, FreqDist, ConditionalFreqDist, RegexpParser, ne_chunk
```

```python
tokens_s1 = word_tokenize(s1)
tokens_s1
```

```python
len(s1), len(tokens_s1)
```

```python
s1_pos_tag = pos_tag(tokens_s1)
s1_pos_tag[:5]
```

```python
s1_tags_freq = FreqDist(t1 for (w1, t1) in s1_pos_tag)
s1_tags_freq
```

```python
s1_tags_freq.plot(50)
```

```python
cdf = ConditionalFreqDist(s1_pos_tag)
lack_fd = cdf['lack']
lack_fd.tabulate()
```

```python
grammar_rule1 = "NP: {<PRP>|<PRP\$>?<NN><IN>?<NN>?<JJ>?}"
cp1 = RegexpParser(grammar_rule1)
result1 = cp1.parse(s1_pos_tag)
display(result1)
```

```python
phrase_en_tokens=word_tokenize(s1)
```

```python
tagged_phrase=pos_tag(phrase_en_tokens)
```

```python
print(ne_chunk(tagged_phrase, binary=True))
```
