## Domain Selection and Dataset Creation

The domain selected for this project is hotel reviews. Specifically, a dataset of reviews for lodgings in Athens city, collected using the Google Places API.

```python
import nltk
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.pyplot as plt

pd.set_option('display.max_colwidth', 15)
```

```python
reviews = pd.read_csv("../datasets/reviews_athens.csv")
reviews.head()
```

```python
reviews_df = reviews.copy()
```

```python
reviews_df.info()
```

**Drop duplicate & missing values (only 1 row will be affected for this dataset):**

```python
reviews_df.drop_duplicates(inplace=True)
reviews_df.dropna(inplace=True)
```

**Convert to *positive* and *negative* sentiment:**

```python
_, ax1 = plt.subplots(nrows=1,ncols=2,figsize=(8,3))

reviews_df.loc[reviews_df.user_rating < 3,'sentiment'] = 'negative'
reviews_df.loc[reviews_df.user_rating > 3,'sentiment'] = 'positive'
reviews_df.drop(index=reviews_df[reviews_df.user_rating==3].index, inplace=True)

t1='Average Rating: {}'.format(round(reviews_df.user_rating.mean(),2))
reviews_df.sentiment.value_counts(True).plot(kind='bar',ax=ax1[0],subplots=True,rot=0)
reviews_df.user_rating.plot(kind='hist',ax=ax1[1],subplots=True)
ax1[0].set_xlabel('')
ax1[0].set_ylabel('%')
ax1[0].set_title('Sentiment Distribution')
ax1[1].set_ylabel('')
ax1[1].set_title('Avg Rating: {}'.format(reviews_df.user_rating.mean().round(1)));
```

**⚠️ Note:** Positive ratings are dominant (more than 80% of the data)

**Randomly drop positive-sentiment rows to have a 1:1 distribution**

```python
np.random.seed(42)

dx = reviews_df.sentiment.value_counts().positive - reviews_df.sentiment.value_counts().negative
random_positives_index = np.random.choice(reviews_df[reviews_df.sentiment=='positive'].index, dx, replace=False)

dataset = reviews_df.drop(index=random_positives_index)
print('Final Dataset:')
dataset[['review_text','sentiment']]
```

**Explore the dataset**

```python
_, ax2 = plt.subplots(nrows=1,ncols=3,figsize=(12,3))

dataset.sentiment.value_counts(True).plot(kind='bar',ax=ax2[0],subplots=True,rot=0)
dataset.review_text.apply(len).plot(kind='hist',ax=ax2[1],subplots=True)
dataset.user_rating.plot(kind='hist',ax=ax2[2],subplots=True)
ax2[0].set_title('Sentiment Distribution')
ax2[0].set_xlabel('')
ax2[0].set_ylabel('%')
ax2[1].set_title('Avg Char Len: {}'.format(dataset.review_text.apply(len).mean().round(1)))
ax2[1].set_ylabel('')
ax2[2].set_ylabel('')
ax2[2].set_title('Avg Rating: {}'.format(reviews_df.user_rating.mean().round(1)));
```

## Preprocessing and Text Analysis

**Perform text cleaning: remove noise, handle emojis, stop words, and other irrelevant tokens.**

```python
def clean(txt):
    txt = re.sub(r'\n|\s\s+',' ',txt).lower()
    return re.sub(r'[^\w\s.,!?]', '', txt)
```

```python
clean('hello  \n  world')
```

```python
dataset.review_text.iloc[1262]
```

```python
dataset['cleaned_review_text'] = dataset.review_text.apply(clean)
dataset.cleaned_review_text.iloc[1262]
```

**Average length after cleaning**

```python
dataset.review_text.apply(len).mean(), '->', dataset.cleaned_review_text.apply(len).mean()
```

**Word cloud of positive and negative reviews**

```python
positive_reviews = ' '.join(dataset[dataset.sentiment == 'positive'].cleaned_review_text)
negative_reviews = ' '.join(dataset[dataset.sentiment == 'negative'].cleaned_review_text)

wordcloud_pos = WordCloud(width=800, height=400).generate(positive_reviews)
wordcloud_neg = WordCloud(width=800, height=400).generate(negative_reviews)

fig, axis = plt.subplots(1,2,figsize=(12,3))
axis[0].imshow(wordcloud_pos)
axis[0].axis('off')
axis[0].set_title('Positive')
axis[1].imshow(wordcloud_neg)
axis[1].axis('off')
axis[1].set_title('Negative');
```
