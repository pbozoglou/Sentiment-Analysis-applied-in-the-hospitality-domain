```python
import nltk
from nltk.tree import Tree
from IPython.display import display
import numpy as np
import pandas as pd
pd.set_option('display.max_colwidth', 15)
```

```python
reviews = pd.read_csv("../datasets/reviews_athens.csv")
reviews.head()
```

#### 1. Preprocessing

```python
reviews.duplicated().sum()
```

```python
reviews.isnull().sum()
```

```python
print('Average Rating:',reviews['user_rating'].mean(),
      '\nAverage review character length',reviews['review_text'].apply(len).mean())
```

```python
reviews['sentiment'] = 'Neutral'
reviews.loc[reviews['user_rating'] < 3,'sentiment'] = 'Negative'
reviews.loc[reviews['user_rating'] > 3,'sentiment'] = 'Positive'
reviews['sentiment'].value_counts(True)
```

```python
#Equalize positive:negative ratio to 1:1 by reducing the positive data at random.

dx = reviews['sentiment'].value_counts()['Positive'] - reviews['sentiment'].value_counts()['Negative']
index = np.random.choice(reviews[reviews['sentiment']=='Positive'].index, dx, replace=False)
reviews_new = reviews.drop(index=index)
reviews_new['sentiment'].value_counts(True)
```

```python
# Drop neutrals
reviews_new.drop(index=reviews_new[reviews_new['sentiment']=='Neutral'].index, inplace=True)
reviews_new['sentiment'].value_counts()
```

```python
dataset = reviews_new[['review_text','sentiment']]
dataset.head(10)
```

```python
# dataset.loc[11,'review_text']
```

```python
print('Average Rating:',reviews_new['user_rating'].mean(),
      '\nAverage review character length',dataset['review_text'].apply(len).mean())
```

#### NLTK Experimentation

```python
from nltk.text import Text
from nltk import FreqDist
```

```python
sentence = dataset.loc[dataset.index[1],'review_text']
example = Text(nltk.word_tokenize(sentence))
example
```

```python
token_sequence = sentence.split()
freq_dist = FreqDist(token_sequence)
freq_dist.most_common(10)
```

```python
example.concordance("was")
```

```python
example.common_contexts(["was"])
```

```python
example.dispersion_plot([i for i in freq_dist][:5])
```
